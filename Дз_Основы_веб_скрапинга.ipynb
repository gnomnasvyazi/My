```python
def search_results(query):
    url = 'https://habr.com/ru/search/?q=python'
    params ={
        'search_query': query
    }
    res = requests.get(url, params)
    soup = BeautifulSoup(res.text)
    refs = soup.find_all('a', 'tm-title__link')

    links_list = []
    for ref in refs:
        links_list.append(ref.get('href'))

    #print(len(links_list))

    links_list = set(links_list)
    #print(len(links_list))

    full_links = list(map(lambda x: 'https://habr.com' + x, links_list))

    return full_links


all_links = search_results('python')
print(all_links)
```

```python
import time
```

```python
def search_results(query, pages):
    url = 'https://habr.com/ru/search/?q=python'
    all_links = []
    params ={
        'search_query': query
    }

    for i in range(1, pages+1):
        params['page'] = i
        res = requests.get(url, params)
        soup = BeautifulSoup(res.text)

        time.sleep(0.2)
        refs = soup.find_all('a', 'tm-title__link')
        links_list = []
        for ref in refs:
            links_list.append(ref.get('href'))

        links_list = set(links_list)

        full_links = list(map(lambda x: 'https://habr.com' + x, links_list))

        all_links += full_links
    return all_links

links = search_results('python', 2)
print(len(links))
print(links)
```

```python
def get_habr(links):
    rows = []
    for link in links:
        soup = BeautifulSoup(requests.get(link).text)
        # print(link)
        time.sleep(0.2)
        date = pd.to_datetime(soup.find('time').get('datetime'))
        title = soup.find('h1', 'tm-title')

        row = {'date': date, 'title': title, 'link': link}
        rows.append(row)
    return pd.DataFrame(rows).reset_index(drop=True)
```

```python
habr = get_habr(links)
habr
```
