```python
import pandas as pd
```

```python
from sqlalchemy import create_engine, text
engine_cloud = create_engine("postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db")
df_city = pd.DataFrame(engine_cloud.connect().execute(text("SELECT * FROM city;")))
```

```python
df_city.head()
```

```python
df_city
```

```python
df_otpusk = pd.read_csv('otpusk.csv', sep=',')
```

```python
df_otpusk
```

```python
df_merged = df_city.merge(df_otpusk, left_on='name', right_on='Города на Английском', how='inner')
```

```python
df_merged
```

```python
df_small = df_merged[['Страна','population']]
```

```python
df_small.head()
```

```python
df_small.groupby('population').mean()
```
