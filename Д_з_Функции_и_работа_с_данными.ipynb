# **Задание 1**
Напишите функцию, которая классифицирует фильмы

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
def class_movies(row):
    if row['rating'] <= 2.0:
        return 'Низкий результат'
    elif  row['rating'] <= 4.0:
        return 'Средний результат'
    else:
        return 'Высокий рейтинг'
```

```python
ratings['class_movies'] = ratings.apply(class_movies, axis=1)
```

```python
ratings[['userId', 'movieId', 'rating', 'timestamp', 'class_movies']]
```

# **Задание 2**
Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
cities = keywords[keywords['keyword'].str.contains('москва|тула|ярославль|петербург|псков|мурманск|владивосток|сахалин|хабаровск', case=False)]
cities
```

```python
def region_identify(row):
    for region, geo_values in geo_data.items():
        for city in geo_values:
            if city in row:
                return region
    return 'undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(region_identify)
keywords
```

```python
my_geo = keywords[keywords['region'].isin(['Северо-Запад', 'Дальний Восток', 'Центр'])]
my_geo
```
