Определите, какому фильму было выставлено больше всего оценок 5.0.

# **Задание** **№2**
Посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
transactions = pd.read_csv('transactions.csv')
transactions.head()
```

```python
power['country'].unique()
```

```python
power[ power['country'].str.contains('lit', case=False) ]['country'].unique()
```

```python
# фильтр на несколько условий сразу
# | - условие ИЛИ
# & AND
# () | (() | () & ())
filtered_countries = power[ (power['country']=='Latvia') | (power['country']=='Lithuania') | (power['country']=='Estonia') & (power['category'] == 4) ]

filtered_countries.head()
```

```python
filtered_countries['country'].unique()
```

```python
pribaltica = power[power['country'].isin(['Latvia', 'Lithuania', 'Estonia'])]
pribaltica.head()
```

```python
pribaltica = pribaltica[pribaltica['category'].isin([4, 12, 21])]
pribaltica.head()
```

```python
pribaltica = pribaltica[(pribaltica['year'] >= 2005) & (pribaltica['year'] <= 2010) & (pribaltica['quantity'] > 0) ]
pribaltica.head(10)
```

```python
sum_quantity = sum(pribaltica['quantity'])
sum_quantity
```
